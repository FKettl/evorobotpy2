```
import numpy as np
import seaborn as sns
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```
experiments = os.listdir("./experiments")
data_frames = []
names = [] 
for x in experiments:
    experiment_name = x
    experiment_name = experiment_name.replace('.csv', '')
    df = pd.read_csv(f"./experiments/{x}")
    df['Experimento'] = [experiment_name for x in range(10)]
    data_frames.append(df)
    
frames = pd.concat(data_frames, axis=0, ignore_index=True)
```

```
lista = list(frames.Experimento.unique())
print(sorted(lista))
```

```
sns.set(rc={"figure.figsize":(12, 9)}) #width=3, #height=4
ax = sns.boxplot(data=frames,
            x="Experiment",
            y="Bestgfit",
            hue="Experimento",
            hue_order=sorted(lista)
            )
#ax.set(xticklabels=[])
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```
