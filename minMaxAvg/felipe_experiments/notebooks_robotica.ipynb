```python
import numpy as np
import seaborn as sns
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
experiments = os.listdir("./results")
data_frames = []
names = [] 
for x in range(len(experiments)):
    experiment_name = experiments[x]
    experiment_name = experiment_name.replace('.csv', '')
    df = pd.read_csv(f"./results/{experiments[x]}")
    df['Experimento'] = [experiment_name for x in range(20)]
    data_frames.append(df)

experiments = os.listdir("./old")
names = [] 
for x in range(len(experiments)):
    experiment_name = experiments[x]
    experiment_name = experiment_name.replace('.csv', '')
    df = pd.read_csv(f"./old/{experiments[x]}")
    df['Experimento'] = [experiment_name for x in range(10)]
    data_frames.append(df)
    
frames = pd.concat(data_frames, axis=0, ignore_index=True)
```

```python
print(list(frames.Experimento.unique()))
```

```python
sns.set(rc={"figure.figsize":(12, 9)}) #width=3, #height=4
ax = sns.boxplot(data=frames,
            x="Experimento",
            y="Bestgfit",
            hue="Experimento",
            width=2
            )
ax.set(xticklabels=[])
plt.show()
```
